# C) Find Top 5 States with Highest Churn Percentage

```python
data.groupby("state")['churn'].mean().sort_values(ascending=False).head(5)*100
```

### *crosstab()*

```python
# Calculating State vs Churn Percentage
# Using crosstab function from pandas library to compute a simple cross-tabulation
# margins = True means to show the sum of both values in a new column 'All'

c = pd.crosstab(data.state, data.churn, margins = True)

c['Percentage_Churn'] = c[True]/(c['All'])*100

print(c.head())
print(type(c))
```

### *info()*

```python
c.info()

# To show the basic information of DataFrame 'c'
```

### *sort_values()*

```python
c.sort_values(by = 'Percentage_Churn', ascending = False).head()

# To sort the records of the dataframe 'c' with respect to the column 'Percentage_Churn' in descending order
```

```python
top_churn_states = c.sort_values(by='Percentage_Churn', ascending=False).head()
top_churn_states
```

```python
# Sort the DataFrame by 'Percentage_Churn' and select the top entries
top_churn_states = c.sort_values(by='Percentage_Churn', ascending=False).head()

# Since 'state' is the index, use it directly in the plot
plt.figure(figsize=(10,5))
bars = plt.bar(top_churn_states.index, top_churn_states['Percentage_Churn'], color='blue')

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate: center of the bar
        bar.get_height(),  # Y-coordinate: top of the bar
        f'{bar.get_height():.1f}%',  # Label text (formatted to 1 decimal place)
        ha='center',  # Center alignment
        va='bottom',  # Place label slightly above the bar
        fontsize=10,
        fontweight='bold')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Percentage Churn')
plt.title('Top States by Percentage Churn')

# Show plot
plt.show()
```

> ## Outcome
> California (CA), New Jersey (NJ), Texas (TX), Maryland (MD), and South Carolina (SC) have the highest percentage churn among all states.
> The churn rate in these states is around 23-26%, which is significantly higher than the overall dataset average.

# Recommendations

**1. Priority Support in High-Churn States:**
- Provide customized discounts or special plans for customers in the high-churn states to reduce churn.
- Study competitors in the high-churn states and match or beat their offers.

**2. Revise International Plans:**
- Offer competitive pricing and packages for international plans to reduce dissatisfaction.
- Conduct surveys or outreach programs to understand why customers with international plans are leaving.

**3. Leverage Voicemail Plans:**
- Offer discounts or incentives to encourage customers to sign up for voicemail plans.
- Educate customers on the benefits of voicemail services.

**4. Improve Customer Support:**
- Identify customers making multiple calls (more than 3 calls) and proactively intervene with personalized support or offers.
- Investigate the reasons for churn among customers who don't contact support. This might involve surveys, feedback analysis, or examining other customer data. Focus on improving overall customer experience beyond just support interactions.
- Provide training for customer service representatives to enhance satisfaction.
